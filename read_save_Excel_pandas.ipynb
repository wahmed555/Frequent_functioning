### 5. Select or Drop Variables
To keep a single variable, you can write in any of the
following three methods -

1. df.ZipCode
2. df[ZipCode]
3. df.loc[:,"ZipCode"]

```python
df.ZipCode
```

```python
df.loc[:,"ZipCode"]
```

### To select variable by column position, you can use df.iloc
function. In the example below, we are selecting second
column. Column Index starts from 0. Hence, 1 refers to
second column.

```python
df.iloc[5:,2]
```

### We can keep multiple variables by specifying desired  variables inside [ ].

1)  df[["TaxRegionName","State"]]

Also, we can make use of df.loc()
function.



```python
df[["TaxRegionName","State"]]
```

```python
df.loc[:,["TaxRegionName","State"]]
```
